# 11. Latent Dirichlet Allocation (LDA)

* El LDA es un ***modelo probabilístico*** que se enmarca dentro de los ***modelos generativos***, ya que trata de describir cómo se crea un documento. (http://www.jmlr.org/papers/volume3/blei03a/blei03a.pdf)


* El LDA propone que ***un documento se crea mediante la selección de los temas y las palabras de acuerdo con las representaciones probabilísticas del texto natural del documento***.


* El LDA calcula **dos matrices de probabilidad P(w|z) y P(z|θ)**, donde:

    - **P(w|z)**: es la probabilidad de que, dado un tema, salga una palabra
    - **P(z|θ)**: es la probabilidad de que un documento pertenezca a un tema


$$ P(w|\theta) = \sum_{z \in  Z} P(w | z) \cdot P(z|\theta) $$


* ***NOTA***: *existe otro modelo llamado PLSI (Probabilistic Latent Semantic Index) que se enmarca dentro de los modelos probabilísticos (al igual que el LDA) y es el enfoque probabilístico del LSI (siguiendo una distribución uniforme).*


* La representación en "*Plate Notation*" del LDA es la siguiente:


<img src="imgs/018_LDA_Plate_Notation.png" style="width: 400px;"/>


* Siendo:

    + ***K***: Número de temas.
    + ***N***: Número de palabras
    + ***M***: Número de documentos
    + ***α***: Parámetro de Dirichlet. Es un vector de K componentes que describe el conocimiento a priori que se tiene sobre cómo se distribuyen los temas en los documentos.
        * Pocos temas -> Valor de α pequeño
        * Muchos temas -> Valor de α grande
    + ***Β***: Parámetro de Dirichlet. Es un vector de N componentes que describe el conocimiento a priori que se tiene sobre cómo se distribuyen las palabras en cada tema.
        * Tema con pocas palabras -> Valor de Β cercano a cero
        * Tema con muchas palabras -> Valor de Β cercano a uno
    + ***θ***: Distribución de probabilidad de que un documento pertenezca a un tema.
    + ***Z***: Distribución de probabilidad de que una palabra pertenezca a un tema.
    + ***W***: Identifica todas las palabras en todos los documentos.
    + ***φ***: Distribución de probabilidad de que, dado un tema, salga una palabra.


* ***NOTA***: Si los parámetros ***α y Β*** de la distribución de Dirichlet son ***iguales a '1'***, el ***LDA se comportaría de la misma manera que el PLSI***, ya que la distribución de Dirichlet con esos parámetros se comportaría como una distribución uniforme.


* Haciendo una analogía con el LSI, el ***LDA nos tiene que proporcionar***:
    - Matriz de probabilidades "***Temas-Palabras***": indica la probabilidad de que, dado un tema, salga una palabra.
    - Matriz de probabilidades "***Documentos-Temas***": indica la probabilidad de que un documento pertenezca a un tema.


* De esta manera podemos ver las relaciones entre palabras y entre documentos.


* ***CUIDADO***: *El LDA trabaja con distribuciones de probabilidad que representan la probabilidad de pertenencia de cada palabra o documento a cada tema. Estas distribuciones de probabilidad no tienen que tratarse como vectores de factores latentes (como en el LSI) para calcular similaridades entre documentos, ya que al tratarse de distribuciones de probabilidad no hay que aplicar medidas de distancia, sino la ***Divergencia de Kullback-Leibler (KL)*** para estudiar las similaridades entre distribuciones de probabilidad.*


$$ KL(p ||q) = \sum_{i}p(i)ln \frac{p(i)}{q(i)} $$

# Ejemplo de LDA con Gensim

* Veamos a continuación un ejemplo sencillo sobre el mismo corpus utilizado en el notebook de LSI, del cual podemos ver que habla de 3 temas deportivos:
    - Fútbol
    - Baloncesto
    - Tenis

````python
import warnings
warnings.filterwarnings('ignore')

# Corpus principal: tres temáticas deportivas (fútbol, baloncesto, tenis)
corpus = [
    # Fútbol
    'gol gol gol messi messi messi messi ronaldo ronaldo liga liga champions champions',
    'futbol futbol futbol gol gol portero portero delantero delantero penalti estadio',
    'gol gol liga liga liga champions champions copa copa messi messi ronaldo ronaldo neymar neymar',
    'penalti penalti arbitro arbitro arbitro arbitro balon balon futbol futbol gol gol estadio estadio',
    # Baloncesto
    'canasta canasta triple triple rebote rebote gol nba nba lebron lebron jordan jordan',
    'baloncesto baloncesto nba nba euroliga euroliga pivot pivot base base alero alero',
    'triple triple canasta canasta lebron jordan jordan gol pabellon pabellon rebote rebote mate mate',
    'nba nba nba lebron lebron canasta canasta triple baloncesto pabellon pabellon euroliga',
    # Tenis
    'tenis tenis nadal nadal djokovic djokovic federer federer gol wimbledon roland_garros saque saque',
    'raqueta raqueta saque saque set set game game nadal pelota pelota gol torneo torneo',
    'wimbledon wimbledon roland_garros roland_garros federer federer djokovic djokovic tenis tenis open open'
]
````

## Creamos el Diccionario y la Matriz (Bolsa de Palabras)

````python
from pprint import pprint
from gensim import corpora
from collections import defaultdict

# Tokenizamos
documents = [doc.split() for doc in corpus]

# Creamos el diccionario (vocabulario)
frequency = defaultdict(int)
for doc in documents:
    for token in doc:
        frequency[token] += 1

dictionary = corpora.Dictionary(documents)
print('Diccionario:')
pprint(dictionary.token2id)

# Creamos la Bolsa de Palabras
corpus_bow = [dictionary.doc2bow(doc) for doc in documents]
print('\nBolsa de Palabras:')
pprint(corpus_bow)
````

## Creamos el Modelo

* Gensim tiene implementado el LDA en la clase ***LdaModel***: https://radimrehurek.com/gensim/models/ldamodel.html


* Como parámetros relevantes necesita:
    1. Corpus
    2. Número de Topics
    3. Diccionario o Vocabulario del Corpus


* ***IMPORTANTE***: *a diferencia del LSI (que se basa en una descomposición matricial determinista, SVD), el LDA es un algoritmo de inferencia iterativo que parte de una inicialización aleatoria. Con corpus muy pequeños, como el de este ejemplo, el resultado puede variar de forma notable según la semilla (`random_state`) y el número de pasadas (`passes`) sobre el corpus. Aquí fijamos `passes=10` para dar más iteraciones al algoritmo y así facilitar la convergencia.*

````python
from gensim.models import LdaModel

lda_model = LdaModel(corpus=corpus_bow, num_topics=3, id2word=dictionary, random_state=10, passes=10)
````

## Matriz de probabilidades "***Documentos-Temas***"

* Obtenemos la probabilidad de que cada documento pertenezca a uno de los 3 temas de la siguiente manera:

````python
import numpy as np

docs_topics = np.array([[tup[1] for tup in lst] for lst in lda_model[corpus_bow]])
docs_topics
````

* Se puede observar que nos devuelve, para cada documento, la probabilidad de que pertenezca a cada tema, y que es un vector de probabilidades ya que la suma de las probabilidades es igual a '1'.


* Para verlo de forma más clara lo mostramos en una tabla:

````python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format

index = ['Doc {}'.format(i + 1) for i, doc in enumerate(documents)]
pd.DataFrame(docs_topics, index=index, columns=['Topic 1', 'Topic 2', 'Topic 3'])
````

## Matriz de probabilidades "***Temas-Palabras***"

* Obtenemos la probabilidad de que, dado uno de los 3 temas, aparezca una de las palabras del vocabulario. Al igual que en la matriz anterior, la suma de todas las probabilidades de las palabras en un tema tiene que sumar '1'.


* A continuación obtenemos la probabilidad de que, dado un tema, aparezca una palabra:

````python
words_topics = lda_model.get_topics()
words_topics
````

* De manera más clara lo mostramos en una tabla:

````python
pd.DataFrame(words_topics, index=['Topic 1', 'Topic 2', 'Topic 3'], columns=dictionary.token2id.keys())
````

## Palabras (términos) más representativas de un tema (topic)

* Dado que podemos obtener la probabilidad de que, dada una palabra (término), esta pertenezca a un tema (topic), podemos obtener las palabras más representativas por topic de la siguiente manera:

````python
for i in range(3):
    print('\nTopic {i}'.format(i=i + 1))
    pprint([dictionary.id2token[term[0]] for term in lda_model.get_topic_terms(i)[0:5]])
````

<hr>

## Topics & Terms

* Gensim nos devuelve una "fórmula" por tema (Topic) que, aplicada a las apariciones de las palabras en los documentos, nos indica la pertenencia del documento a ese tema. El que mayor valor tenga tras aplicar la fórmula del tema al documento significará que tiene mayor propensión a pertenecer a ese tema.


* Si os fijáis, esa fórmula la construye como el sumatorio de la aparición de la palabra en el documento, multiplicado por la probabilidad de que en ese tema aparezca esa palabra:

````python
lda_model.print_topics(num_words=10)
````

## Conclusiones

* Con `random_state=10` y `passes=10`, el LDA separa muy limpiamente los tres temas: los documentos de fútbol (1-4) obtienen una probabilidad de en torno al 0.95 en el Topic 1, los de baloncesto (5-8) en el Topic 3, y los de tenis (9-11) en el Topic 2, sin apenas mezcla entre documentos de temas distintos.


* La palabra "gol" está presente en documentos de los tres temas de forma intencionada. En las probabilidades "Temas-Palabras" esto se traduce en que "gol" aparece con un peso relevante en los tres topics, en lugar de estar asociada de forma exclusiva a uno solo. Es la manera que tiene el modelo probabilístico de reflejar que un término es poco discriminativo, aunque en este caso no ha sido suficiente para impedir una separación limpia de los documentos.


* Esta separación tan clara no debe interpretarse como una garantía general del LDA. Al ser un algoritmo de inferencia iterativo con inicialización aleatoria, el resultado depende de la semilla y del número de pasadas: con otras semillas (probarlo cambiando `random_state`) el reparto de documentos entre topics puede ser mucho más confuso, incluso mezclando documentos de temas distintos en un mismo topic. Esta es una diferencia importante frente al LSI, cuya descomposición (SVD) es determinista.


* A diferencia del LSI, aquí no podemos usar una distancia geométrica para comparar documentos o palabras, ya que estamos comparando distribuciones de probabilidad. Para ello se emplea la Divergencia de Kullback-Leibler.
